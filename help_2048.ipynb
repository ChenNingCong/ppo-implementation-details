```python
import gymnasium as gym 

%load_ext autoreload
%autoreload 3
from ppo_agent_2048 import *
num_envs = 4
env = gym.make_vec("gymnasium_2048:gymnasium_2048/TwentyFortyEight-v0", num_envs=num_envs)
agent = Agent(envs=env, dropout=0)
agent
trainer = PPOTrainer(
    env =env,
    agent = agent,
    rollout_num = 100000,
    rollout_length = 128,
    epsilon = 0.2,
    value_loss_weight = 0.2,
    clip_norm = 0.5,
    norm_adv = True,
    ppo_epoch = 4,
    ppo_num_minibatch = 4,
    gamma = 0.99,
    use_gae = True,
    lam = 0.95
)
def wrapper():
    trainer.train()
wrapper()
```

```python
env =gym.wrappers.RecordVideo(gym.make("gymnasium_2048:gymnasium_2048/TwentyFortyEight-v0", render_mode="rgb_array"), f"video/1.mp4")
@torch.no_grad
def validate(self, new_env):
    # this returns undiscounted reward
    old_env = self.env
    try:
        self.env = new_env  # set new environment
        obs, _ = self.env.reset()
        # add a batch dim
        obs = torch.tensor(obs, device=self.device).unsqueeze(0)
        rewards = []
        with torch.no_grad():
            self.agent.eval()
            total_reward = 0
            terminated = truncated = False 
            while not (terminated or truncated):
                action, action_log_prob, entropy, state_value = self.agent.get_action_and_value(obs, action=None)
                next_obs, reward, terminated, truncated, _ = self.env.step(action.item())
                # # we disable truncation...
                # assert not np.any(truncated)
                total_reward += reward
                obs = next_obs
                obs = torch.tensor(obs, device=self.device).unsqueeze(0) 
    finally:
        self.env.close()
        self.env = old_env
    return total_reward
validate(trainer, env)
```
